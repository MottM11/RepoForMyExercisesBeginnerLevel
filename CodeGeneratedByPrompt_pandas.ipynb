```python
# prompt: Provide me some sample coding using pandas

import pandas as pd

# Create a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 22, 28],
        'City': ['New York', 'London', 'Paris', 'Tokyo']}
df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Accessing columns
print("\nNames:")
print(df['Name'])

# Accessing rows by index
print("\nRow at index 1:")
print(df.loc[1])

# Filtering data
print("\nPeople older than 25:")
print(df[df['Age'] > 25])

# Grouping and aggregation
print("\nAverage age by city:")
print(df.groupby('City')['Age'].mean())

# Sorting data
print("\nSorted by age:")
print(df.sort_values('Age'))

# Adding a new column
df['Country'] = ['USA', 'UK', 'France', 'Japan']
print("\nDataFrame with new column:")
print(df)

# Removing a column
df = df.drop('Country', axis=1)
print("\nDataFrame with column removed:")
print(df)

# Handling missing data
data_with_missing = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
                     'Age': [25, None, 22, 28],
                     'City': ['New York', 'London', 'Paris', 'Tokyo']}
df_missing = pd.DataFrame(data_with_missing)

print("\nDataFrame with missing data:")
print(df_missing)

# Fill missing values with mean
df_missing['Age'] = df_missing['Age'].fillna(df_missing['Age'].mean())
print("\nDataFrame with missing values filled:")
df_missing

```
